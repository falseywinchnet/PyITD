```python
import numpy as np
import matplotlib.pyplot as plt

def custom_filter_engine(x,offset=0):
    out = 0.5 * x
    residual = 0.5 * x
    grad = np.gradient(x)
    a = 1- 0.36787944

    # First-order correction
    flip = np.sign(grad[:-1]) != np.sign(grad[1:])
    flip = np.concatenate([[False], flip])  # align with x length
    delta = a * residual
    out += np.where(flip, delta, -delta)
    residual *= 0.36787944

    # Higher-order corrections (2nd to 5th)
    for k in range(2, 9):
        grad = np.gradient(residual)
        flip = np.sign(grad[:-1]) != np.sign(grad[1:])
        flip = np.concatenate([[False], flip])  # maintain length
        delta = a * residual
        out += np.where(flip, delta, -delta)
        residual *= 0.36787944
    return out

# Test signal
x = np.linspace(-10, 10, 48000)
signal = np.sin(x) + 0.3 * np.cos(5 * x)
signal = np.sin(x) + 0.44 * np.cos(7 * x)



# Apply custom filter
filtered_signal = custom_filter_engine(signal)
filtered_signal2 = custom_filter_engine(filtered_signal)

# Plot results
plt.plot(x, signal, label='Original')
plt.plot(x, filtered_signal, label='Filtered')
plt.plot(x, filtered_signal2, label='Filtered2')
plt.legend()
plt.title("Custom Filter Engine Output")
plt.show()
```

```python
from scipy.interpolate import CubicSpline

def extract_trend(x, signal):
    f1 = custom_filter_engine(signal)
    f2 = custom_filter_engine(f1)
    d2 = np.gradient(np.gradient(f2))

    zero_crossings = np.where(np.diff(np.sign(d2)))[0]

    # Filter closely spaced crossings to avoid oversampling

    # Ensure endpoints are included
    zero_crossings = np.unique(np.concatenate([[0], zero_crossings, [len(x)-1]]))

    cs = CubicSpline(x[zero_crossings], signal[zero_crossings], bc_type='natural')
    trend = cs(x)
    return trend, zero_crossings
```

```python
def decompose_signal(x, signal, max_iter=10, tol=1e-6):
    components = []
    residual = signal.copy()

    for i in range(max_iter):
        trend, knots = extract_trend(x, residual)
        components.append(trend)
        new_residual = residual - trend

        if np.linalg.norm(new_residual - residual) < tol:
            break
        residual = new_residual

    return components, residual
```

```python
components, residual = decompose_signal(x, signal)
```

```python
for component in components:
    plt.plot(x, component)
    plt.show()
```
